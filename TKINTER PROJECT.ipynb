```python
from tkinter import *
from tkinter import messagebox
def del_task():
    messagebox.showwarning("NOTICE!!!", "Do you want to delete task?")
    box.delete(ANCHOR)
    
def add_task():
    task=my_entry.get()
    if task != "":
        box.insert(END, task)
        my_entry.delete(0, END)
        
def completed():
    box.itemconfig(box.curselection(), fg="red")
    box.selection_clear(0, END)
    
root=Tk()
root.config(bg="green")

my_frame=Frame(root,bg="green")
my_frame.grid(row=0,column=0)

my_frame1=Frame(root,bg="green")
my_frame1.grid(row=0, column=1)

box=Listbox(my_frame, width=50, 
            height=10,
            font=("bold",12),
            selectbackground="green",
            activestyle="none",
            bg="white")
box.grid(row=0, column=0)

my_entry=Entry(my_frame)
my_entry.grid(row=1, column=0)


my_button1=Button(my_frame1, text="Add Task", command=add_task)
my_button1.grid(row=0, column=0)

my_button2=Button(my_frame1, text="Delete Task", command=del_task)
my_button2.grid(row=1, column=0)

my_button3=Button(my_frame1, text="Mark As Completed", command=completed)
my_button3.grid(row=2, column=0)



list1=["name", "age", "date of birth", "yearmof birth"]
for task in list1:
    box.insert(END,task)



root.mainloop()
```
